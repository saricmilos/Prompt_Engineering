# Prompt Formatting and Structure Tutorial

## Overview

## Motivation

The way a prompt is written is not merely cosmetic. It fundamentally shapes how an AI model interprets a request, what assumptions it makes about context and audience, and how it organizes its response. A vague or poorly structured prompt often produces generic, unfocused, or inconsistent answers, while a well-crafted one can guide the model toward precise, useful, and contextually appropriate output.

This distinction becomes especially critical in production applications — customer-facing chatbots, automated content pipelines, coding assistants, educational tools — where response quality and predictability are non-negotiable. Mastering prompt structure allows engineers, researchers, and practitioners to reduce prompt iteration time, improve reliability, and extract far more value from AI systems without changing the underlying model at all.

This tutorial bridges the gap between theoretical prompt engineering principles and practical, replicable techniques you can apply immediately.

## Key Components

1. **Different prompt formats (Q&A, dialogue, instructions)** — We'll examine how framing a prompt as a question, a conversation, or a directive affects the tone, depth, and structure of the model's response. Each format is suited to different tasks and audiences.

2. **Structural elements (headings, bullet points, numbered lists)** — We'll explore how embedding formatting cues directly inside the prompt — such as asking the model to respond "in bullet points" or "with a numbered list of steps" — primes it to organize information more clearly and predictably.

3. **Comparison of prompt effectiveness** — Rather than simply presenting formats in isolation, we'll run side-by-side comparisons using a consistent topic, so differences in output can be directly attributed to differences in prompt structure rather than subject matter.

4. **Best practices for prompt formatting** — Drawing from our experiments, we'll synthesize a set of actionable guidelines: when to use which format, how to use whitespace and punctuation strategically, when explicit instructions outperform implicit ones, and common pitfalls to avoid.

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from typing import Iterator
from langchain_core.messages import BaseMessage
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
```

```python
# 1. Load the .env file
# This automatically puts variables into os.environ
load_dotenv() 

# 2. Check if the key exists (Fail Fast)
if not os.getenv("OPENAI_API_KEY"):
    raise ValueError("OPENAI_API_KEY is missing! Check your .env file.")

# 3. Initialize (LangChain automatically looks for the env var)
llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
def get_response(prompt: str) -> str:
    """
    Helper function to stream model response for better UX 
    and return the full text.
    """
    full_response = ""
    
    # .stream() is the modern standard for real-time output
    for chunk in llm.stream(prompt):
        content = chunk.content
        print(content, end="", flush=True)
        full_response += content
        
    print(f"\n{'-' * 50}")
    return full_response
```

## Exploring Different Prompt Formats

Let's explore various prompt formats using the topic of photosynthesis as our consistent theme.

### 1. Question and Answer (Q&A) Format

```python
qa_prompt = """Q: What is photosynthesis?
A:"""

get_response(qa_prompt)
```

### 2. Dialogue Format

```python
messages = [
    SystemMessage(content="You are a helpful and concise biology teacher."),
    HumanMessage(content="Can you explain photosynthesis to me?"),
    AIMessage(content="Certainly! Photosynthesis is the process plants use to turn sunlight into energy."),
    HumanMessage(content="What does a plant need for photosynthesis?")
]

response = get_response(messages)
```

### 3. Instruction Format

```python
instruction_prompt = """Provide a brief explanation of photosynthesis, including its main components and importance."""

get_response(instruction_prompt)
```

## Impact of Structural Elements

Now, let's examine how structural elements like headings and lists affect the AI's response.

### 1. Using Headings

```python
headings_prompt = """Explain photosynthesis using the following structure:

# Definition

# Process

# Importance
"""

get_response(headings_prompt)
```

### 2. Using Bullet Points

```python
bullet_points_prompt = """List the key components needed for photosynthesis:

• 
• 
• 
"""

get_response(bullet_points_prompt)
```

### 3. Using Numbered Lists

```python
numbered_list_prompt = """Describe the steps of photosynthesis in order:

1.
2.
3.
4.
"""

get_response(numbered_list_prompt)
```

## Comparing Prompt Effectiveness

Let's compare the effectiveness of different prompt structures for a specific task.

```python
# Modern approach using a dictionary and f-string formatting
prompts = {
    "Zero-shot": "Explain the importance of photosynthesis for life on Earth.",
    "Structured": (
        "Explain the importance of photosynthesis for life on Earth. "
        "Structure your answer as follows:\n1. Oxygen production\n"
        "2. Food chain support\n3. Carbon dioxide absorption"
    ),
    "Few-shot/Completion": (
        "Q: Why is photosynthesis important for life on Earth?\n"
        "A: Photosynthesis is crucial for life on Earth because:\n1. \n2. \n3."
    )
}

for label, prompt in prompts.items():
    print(f"\n{'='*10} Testing: {label} {'='*10}")
    # Assuming get_response is your LLM call function
    response = get_response(prompt)
    print(response)
```
